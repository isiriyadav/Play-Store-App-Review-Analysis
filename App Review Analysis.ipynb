###Data Cleaning

````python
def convert_float(a):
  a = str(a)
  for i in a:
    if i in ',+Mk$':
      if i == 'k':
        a = str(a).replace(i,'')
        a = eval(a)/1024.0
      a = str(a).replace(i,'')
  if a == 'Free' or a =='Varies with device'or a =='Everyone':
    a=0
  else:
    a=eval(str(a))
  return a
````

````python
def convert_Million(a):
  if 'M' in a:
    a = str(a).replace('M', '')
    a = eval(str(a)) * 1000000
  return a
````

````python
#convert installs,price,reviews to float value so that it better for calculation
new_ps_data['Installs'] = new_ps_data['Installs'].apply(convert_float)
new_ps_data['Price'] = new_ps_data['Price'].apply(convert_float)
new_ps_data['Size'] = new_ps_data['Size'].apply(convert_float)
````

````python
def extract_year(str):
  from datetime import datetime
  # Input date string
  date_string = str
  # Convert the date string to a datetime object
  date_object = datetime.strptime(date_string, '%B %d, %Y')
  # Extract the year
  year = date_object.year
  return year
````

````python
updates = new_ps_data1['Last Updated'].apply(extract_year)
updates.dropna()
updates.value_counts()
````

````python
plt.figure(figsize = (12,8))
sns.barplot(x = updates.value_counts().index, y = updates.value_counts(), alpha=0.9)
plt.xlabel("Last updated")
plt.title("The last updates that is done to apps")
plt.ylabel("Number of apps")
plt.show()
````
